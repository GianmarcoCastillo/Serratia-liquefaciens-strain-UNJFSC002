```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_3 = pd.read_csv("/home/marcos/Escritorio/SERRATIA/phenotiping/CHINO100.csv", na_values="-", index_col=0)
```

```python
fig, ax = plt.subplots(figsize=(9, 3.8))
sns.heatmap(df_3, cmap='coolwarm', vmin=0, vmax=5, linewidths=0.4, linecolor='black')
plt.xlabel("Genes", fontsize=8)
plt.ylabel("Scientific names")
plt.yticks(fontstyle='italic', fontsize=10)
plt.title("Nitrogen Fixation")
plt.savefig("Nitrogen_Fixation_FINISH010.pdf", format="pdf", dpi=600, bbox_inches="tight")
plt.xticks(rotation=90)
plt.show()
```
